```python
from vncorenlp import VnCoreNLP
rdrsegmenter = VnCoreNLP("/content/drive/My Drive/BERT/SA/vncorenlp/VnCoreNLP-1.1.1.jar", annotators="wseg", max_heap_size='-Xmx500m') 
```

```python
import re

train_path = '/content/drive/My Drive/BERT/SA/13-categories-train.csv'
test_path = '/content/drive/My Drive/BERT/SA/13-categories-test.csv'
```

```python
data_train = pd.read_csv(train_path)

train_text = []
i = 0
for sample in data_train['content_text']:
  i+=1
  text = rdrsegmenter.tokenize(sample)
  text = ' '.join([' '.join(x) for x in text])
  train_text.append(text)
  print(i)
```

```python
data_train_final = {'content' : train_text,
                    'label' : data_train['original_label']}
```

```python
df_final = DataFrame(data_train_final, columns = ['content', 'label'])
df_final.to_csv('VNCore_train_data_13.csv',index=None, header=True)
```

```python
print(len(df_final))
```

```python
data_test = pd.read_csv(test_path)

test_text = []
i = 0
for sample in data_test['content_text']:
  i+=1
  text = rdrsegmenter.tokenize(sample)
  text = ' '.join([' '.join(x) for x in text])
  test_text.append(text)
  print(i)
```

```python
data_test_final = {'content' : test_text,
                    'label' : data_test['original_label']}

```

```python
df_test_final = DataFrame(data_test_final, columns = ['content', 'label'])
df_test_final = df_test_final[df_test_final['label'] != 'thoi-su']
```

```python
print(len(df_test_final))
```

```python
df_test_final.to_csv('VNCore_test_data_13.csv',index=None, header=True)
```
